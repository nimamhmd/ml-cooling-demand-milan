```python
plt.figure(figsize=(15,5))
plt.plot(era5["temperature_2m"], linewidth=0.7)
plt.title("Daily 2m Temperature in Milan (1990–2024)")
plt.xlabel("Year")
plt.ylabel("Temperature (°C)")
plt.grid(True)
plt.show()
```

```python
# Compute annual mean temperatures
annual_mean_temp = era5["temperature_2m"].resample("Y").mean()

annual_mean_temp.head()
```

```python
plt.figure(figsize=(12,5))
plt.plot(annual_mean_temp.index.year, annual_mean_temp.values, marker='o')
plt.title("Annual Mean Temperature in Milan (1990–2024)")
plt.xlabel("Year")
plt.ylabel("Mean Temperature (°C)")
plt.grid(True)
plt.show()
```

```python
from scipy.stats import linregress

years = annual_mean_temp.index.year
temps = annual_mean_temp.values

slope, intercept, r, p, se = linregress(years, temps)

warming_per_decade = slope * 10

slope, warming_per_decade, r, p
```

```python
plt.figure(figsize=(12,5))
plt.plot(years, temps, marker='o', label="Annual Mean Temperature")
plt.plot(years, intercept + slope * years, color='red', label="Trend Line")
plt.title("Warming Trend in Milan (1990–2024)")
plt.xlabel("Year")
plt.ylabel("Mean Temperature (°C)")
plt.grid(True)
plt.legend()
plt.show()
```

```python
baseline_period = annual_mean_temp["1991":"2020"]
baseline_mean = baseline_period.mean()

baseline_mean
```

```python
anomalies = annual_mean_temp - baseline_mean
anomalies.head()
```

```python
plt.figure(figsize=(12,5))
plt.axhline(0, color='black', linewidth=1)  # zero line
plt.bar(anomalies.index.year, anomalies.values, color=['red' if x > 0 else 'blue' for x in anomalies.values])
plt.title("Annual Temperature Anomalies in Milan (Baseline: 1991–2020)")
plt.xlabel("Year")
plt.ylabel("Temperature Anomaly (°C)")
plt.grid(True)
plt.show()
```

```python
# Compute monthly climatology across all years
monthly_climatology = era5["temperature_2m"].groupby(era5.index.month).mean()

monthly_climatology
```

```python
plt.figure(figsize=(12,5))
plt.plot(range(1,13), monthly_climatology.values, marker='o')
plt.title("Monthly Climatology of 2m Temperature in Milan (1990–2024)")
plt.xlabel("Month")
plt.ylabel("Mean Temperature (°C)")
plt.xticks(range(1,13))
plt.grid(True)
plt.show()
```

```python
early = era5.loc["1990":"2005"]["temperature_2m"]
late  = era5.loc["2006":"2024"]["temperature_2m"]

early_clim = early.groupby(early.index.month).mean()
late_clim  = late.groupby(late.index.month).mean()

early_clim, late_clim
```

```python
plt.figure(figsize=(12,5))
plt.plot(range(1,13), early_clim.values, marker='o', label="1990–2005")
plt.plot(range(1,13), late_clim.values, marker='o', label="2006–2024")
plt.title("Seasonal Warming: Early vs Late Periods in Milan")
plt.xlabel("Month")
plt.ylabel("Mean Temperature (°C)")
plt.xticks(range(1,13))
plt.grid(True)
plt.legend()
plt.show()
```

```python
climate_vars = [
    "dewpoint_temperature_2m",
    "total_precipitation",
    "volumetric_soil_water_layer_1",
    "u_component_of_wind_10m",
    "v_component_of_wind_10m",
    "wind_speed_10m",
    "surface_net_solar_radiation",
    "surface_solar_radiation_downwards",
    "surface_net_thermal_radiation",
    "surface_thermal_radiation_downwards"
]

era5[climate_vars].describe().T
```

```python
monthly_clim = era5.groupby(era5.index.month)[climate_vars].mean()
monthly_clim
```

```python
from scipy.stats import linregress

trends = {}

years = era5.index.year

for var in climate_vars:
    vals = era5[var].values
    slope, intercept, r, p, se = linregress(years, vals)
    trends[var] = {
        "slope_per_year": slope,
        "trend_per_decade": slope * 10,
        "r": r,
        "p": p
    }

trends
```

```python
era5_corr = era5[["temperature_2m"] + climate_vars].corr()
era5_corr["temperature_2m"]
```

```python
import numpy as np

era5["wind_speed_mag"] = np.sqrt(
    era5["u_component_of_wind_10m"]**2 + era5["v_component_of_wind_10m"]**2
)
```

```python
era5["log_precip"] = np.log1p(era5["total_precipitation"])
era5["log_solar_down"] = np.log1p(era5["surface_solar_radiation_downwards"])
era5["log_solar_net"] = np.log1p(era5["surface_net_solar_radiation"])
era5["log_thermal_down"] = np.log1p(era5["surface_thermal_radiation_downwards"])
era5["log_thermal_net"] = np.log1p(era5["surface_net_thermal_radiation"] - era5["surface_net_thermal_radiation"].min() + 1)
```

```python
import matplotlib.pyplot as plt

variables = [
    "temperature_2m",
    "dewpoint_temperature_2m",
    "wind_speed_mag",
    "volumetric_soil_water_layer_1",
    "log_precip",
    "log_solar_down",
    "log_solar_net",
    "log_thermal_down",
    "log_thermal_net"
]

plt.figure(figsize=(16, 14))

for i, var in enumerate(variables, 1):
    plt.subplot(3, 3, i)
    plt.hist(era5[var], bins=40, color="skyblue", edgecolor="black")
    plt.title(f"Histogram of {var}")
    plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
from scipy.stats import gaussian_kde
import numpy as np

plt.figure(figsize=(16, 14))

for i, var in enumerate(variables, 1):
    data = era5[var].dropna()
    kde = gaussian_kde(data)

    x_vals = np.linspace(data.min(), data.max(), 300)

    plt.subplot(3, 3, i)
    plt.plot(x_vals, kde(x_vals), linewidth=2)
    plt.title(f"Density Curve: {var}")
    plt.grid(True)

plt.tight_layout()
plt.show()
```
