```python
def quick_excel_inspect(path, nrows=5):
    print("\n" + "="*80)
    print(f"FILE: {path}")
    print("="*80)

    try:
        xls = pd.ExcelFile(path)
    except Exception as e:
        print("Error opening file:", e)
        return
    
    print("Available sheets:")
    for sheet in xls.sheet_names:
        print(f"  - {sheet}")

    for sheet in xls.sheet_names:
        print("\n" + "-"*40)
        print(f"SHEET: {sheet}")
        print("-"*40)

        try:
            df = pd.read_excel(xls, sheet_name=sheet, nrows=nrows)
            print(f"Preview shape: {df.shape}")
            print("Columns:", list(df.columns))
            print(df.head())
        except Exception as e:
            print("Error reading sheet:", e)
```

```python
import pandas as pd
import csv

def detect_delimiter(path, sample_size=50000):
    """Detect the delimiter of a CSV file using csv.Sniffer."""
    with open(path, "r", encoding="latin1") as f:  # latin1 is tolerant for weird chars
        sample = f.read(sample_size)
    sniffer = csv.Sniffer()
    dialect = sniffer.sniff(sample, delimiters=[",", ";", "\t", "|"])
    return dialect.delimiter

def quick_csv_inspect(path, nrows=5):
    print("\n" + "="*80)
    print(f"FILE: {path}")
    print("="*80)

    try:
        delim = detect_delimiter(path)
        print(f"Detected delimiter: {repr(delim)}")
    except Exception as e:
        print("Could not detect delimiter, defaulting to comma. Error:", e)
        delim = ","

    try:
        df = pd.read_csv(path, sep=delim, nrows=nrows, low_memory=False)
    except Exception as e:
        print("Error reading CSV:", e)
        return

    print(f"Preview shape (first {nrows} rows): {df.shape}")
    print("Columns:")
    print(list(df.columns))
    print("\nHead:")
    print(df.head())
```

```python
for path in csv_files:
    quick_csv_inspect(path, nrows=5)
```

```python
# ============================================
# 10.7.8.2 — Load EPC dataset & inspect
# ============================================

epc = pd.read_csv(epc_path, sep=",", low_memory=False)

print("Raw EPC shape:", epc.shape)

epc.head()
```

```python
# ============================================
# 10.7.8.3 — Filter EPC dataset to MILAN only
# ============================================

epc_milan = epc[epc["COMUNE_CATASTALE"] == "MILANO"].copy()

print("Milan EPC shape:", epc_milan.shape)

epc_milan.head()
```

```python
# ============================================
# 10.7.8.4 — Select core ML + merging columns
# ============================================

core_cols = [
    # Identification / address
    "COD_APE",
    "DATA_INS",
    "COMUNE",
    "INDIRIZZO",

    # Building characteristics
    "ANNO_COSTRUZIONE",

    # ML target variables
    "EP_GL_NREN",
    "EP_GL_REN",
    "CLASSE_ENERGETICA",

    # Cooling-related predictors
    "CLIMATIZZAZIONE_ESTIVA",
    "SUPERF_UTILE_RAFFRESCATA",
    "VOLUME_LORDO_RAFFRESCATO",
    "CONSUMI_TELERAFFRESCAMENTO",
]

epc_clean = epc_milan[core_cols].copy()

print("Clean EPC shape:", epc_clean.shape)
epc_clean.head()
```

```python
# ============================================
# 10.7.8.5 — Basic Cleaning (dates + numeric + missing values)
# ============================================

# 1. Convert DATA_INS to datetime (day-first format)
epc_clean["DATA_INS"] = pd.to_datetime(epc_clean["DATA_INS"], errors="coerce", dayfirst=True)

# 2. Convert numeric columns
numeric_cols = [
    "EP_GL_NREN",
    "EP_GL_REN",
    "SUPERF_UTILE_RAFFRESCATA",
    "VOLUME_LORDO_RAFFRESCATO",
    "CONSUMI_TELERAFFRESCAMENTO",
]

for col in numeric_cols:
    epc_clean[col] = pd.to_numeric(epc_clean[col], errors="coerce")

# 3. Show missing values summary
missing_summary = epc_clean.isna().sum()

missing_summary
```

```python
# ============================================
# 10.7.8.5 — Basic Cleaning (dates + numeric + missing values)
# ============================================

# 1. Convert DATA_INS to datetime (day-first format)
epc_clean["DATA_INS"] = pd.to_datetime(epc_clean["DATA_INS"], errors="coerce", dayfirst=True)

# 2. Convert numeric columns
numeric_cols = [
    "EP_GL_NREN",
    "EP_GL_REN",
    "SUPERF_UTILE_RAFFRESCATA",
    "VOLUME_LORDO_RAFFRESCATO",
    "CONSUMI_TELERAFFRESCAMENTO",
]

for col in numeric_cols:
    epc_clean[col] = pd.to_numeric(epc_clean[col], errors="coerce")

# 3. Show missing values summary
missing_summary = epc_clean.isna().sum()

missing_summary
```

```python
# ============================================
# 10.7.8.6 — Normalize address field
# ============================================

# Work on a copy of the address string
addr = epc_clean["INDIRIZZO"].astype(str)

# 1) Uppercase
addr = addr.str.upper()

# 2) Strip leading/trailing spaces
addr = addr.str.strip()

# 3) Collapse multiple spaces into one
addr = addr.str.replace(r"\s+", " ", regex=True)

# Save as new column
epc_clean["INDIRIZZO_NORM"] = addr

print("Example normalized addresses:")
print(epc_clean[["INDIRIZZO", "INDIRIZZO_NORM"]].head())
```

```python
# ============================================
# 10.7.9.1 — Handle missing values
# ============================================

# Only 2673 missing values exist and they are all in CONSUMI_TELERAFFRESCAMENTO
epc_clean["CONSUMI_TELERAFFRESCAMENTO"] = epc_clean["CONSUMI_TELERAFFRESCAMENTO"].fillna(0)

# Verify again
epc_clean.isna().sum()
```

```python
# ============================================
# 10.7.9.2 — Clean ANNO_COSTRUZIONE
# Convert ranges like "1993-2006" to 1993
# ============================================

def clean_year(x):
    if pd.isna(x):
        return None
    x = str(x)
    # If a range, take the first part
    if "-" in x:
        return int(x.split("-")[0])
    # Otherwise convert directly
    try:
        return int(x)
    except:
        return None

epc_clean["ANNO_COSTRUZIONE_CLEAN"] = epc_clean["ANNO_COSTRUZIONE"].apply(clean_year)

# Preview
epc_clean[["ANNO_COSTRUZIONE", "ANNO_COSTRUZIONE_CLEAN"]].head(10)
```

```python
# ============================================
# 10.7.9.3 — Final EPC cleaning
# ============================================

# 1. Fill missing construction years with the median of valid years
median_year = epc_clean["ANNO_COSTRUZIONE_CLEAN"].median()
epc_clean["ANNO_COSTRUZIONE_CLEAN"] = epc_clean["ANNO_COSTRUZIONE_CLEAN"].fillna(median_year)

# Convert to integer
epc_clean["ANNO_COSTRUZIONE_CLEAN"] = epc_clean["ANNO_COSTRUZIONE_CLEAN"].astype(int)

# 2. Build final dataset
epc_clean_final = epc_clean[[
    "COD_APE", "DATA_INS", "COMUNE", "INDIRIZZO", "INDIRIZZO_NORM",
    "ANNO_COSTRUZIONE_CLEAN", 
    "EP_GL_NREN", "EP_GL_REN", "CLASSE_ENERGETICA",
    "CLIMATIZZAZIONE_ESTIVA",
    "SUPERF_UTILE_RAFFRESCATA", "VOLUME_LORDO_RAFFRESCATO",
    "CONSUMI_TELERAFFRESCAMENTO"
]]

print("Final EPC dataset shape:", epc_clean_final.shape)
epc_clean_final.head()
```

```python
# ============================================
# 10.7.9.4 — Save EPC clean dataset
# ============================================

save_path = r"F:\Building Engineering - Polytechnic University of Turin\Masters Thesis\23-11-2025\Data Collections\CENED\epc_clean_final.csv"
epc_clean_final.to_csv(save_path, index=False)

save_path
```
